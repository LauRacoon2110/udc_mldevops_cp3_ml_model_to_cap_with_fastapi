```python
import pandas as pd
from pathlib import Path
import os

import numpy as np


from ydata_profiling import ProfileReport
```

```python
# Get the root path of the working directory
root_path = Path(os.getcwd()).resolve().parent
data_path = Path(root_path, 'data', 'census.csv')
# print(str(data_path))

census_raw_df = pd.read_csv(data_path)
```

```python
census_cls_df = census_raw_df.copy()
census_cls_df = census_cls_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
census_cls_df.replace('?', np.nan, inplace=True)

census_cls_df = census_cls_df.drop_duplicates()
```

```python
census_cls_df.head(n=20)
```

```python
profile = ProfileReport(
    census_cls_df,
    title="Census (PREP) Data Profiling Report",
    explorative=True,        # Enables correlations, missing values, etc.
    minimal=False,           # Set to True for faster runs
    # correlations={"pearson": True, "spearman": True},
    # missing_diagrams={"bar": True, "matrix": True, "heatmap": True},
    # duplicates={"head": 10}, # Shows sample of duplicate rows
)

# Display in notebook
profile.to_notebook_iframe()

# Save it to an HTML file
profile.to_file("census_prep_data_profiling_report.html")
```

```python
# due to <1% of missing values in the dataset, we can drop the rows with missing values
census_cls_df = census_cls_df.dropna()
```

```python
# save new dataset to csv
census_cls_df.to_csv(Path(root_path, 'data', 'census_cls.csv'), index=False)
```
